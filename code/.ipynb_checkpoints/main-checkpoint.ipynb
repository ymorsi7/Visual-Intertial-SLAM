```python
import numpy as np
from pr3_utils import *
import matplotlib.pyplot as plt
from tqdm import tqdm
from main_utils import *
import cv2
from feature_detection import *

filename = "../data/dataset00/dataset00.npy"
v_t, w_t, timestamps, features, K_l, K_r, extL_T_imu, extR_T_imu = load_data(filename)
```

## 1) IMU localization via EKF prediction

```python
dt = np.diff(timestamps, axis=0) # time steps
N = len(dt)

poses = np.zeros((N + 1, 4, 4)) # init
poses[0] = np.eye(4)

for i in range(N):
    v = v_t[i]  # linear v
    w = w_t[i]  # ang v
    
    xi = np.zeros((4, 4)) # twist mat
    xi[:3, :3] = axangle2skew(w)
    xi[:3, 3] = v
    
    T_update = twist2pose(xi * dt[i]) # pose updating
    poses[i + 1] = poses[i] @ T_update
```

```python
fig, ax = visualize_trajectory_2d(poses, path_name="IMU Trajectory", show_ori=True)
plt.xlabel('X (meters)')
plt.ylabel('Y (meters)')
plt.title('IMU Trajectory')
plt.show()
```

## 3) Landmark Mapping via EKF Update

```python
landmark_map = landmark_mapping(v_t, w_t, timestamps, features, K_l, K_r, extL_T_imu, extR_T_imu, poses) 
```

```python
plt.figure()
good_points = np.any(landmark_map != 0, axis=0)
plt.scatter(landmark_map[0, good_points], landmark_map[1, good_points], s=3, marker='.', alpha=0.7, label='Landmarks')
plt.plot(poses[:, 0, 3], poses[:, 1, 3], 'r-', linewidth=3, label='IMU Trajectory')
plt.axis('equal')
plt.title('EKF Mapping')
plt.xlabel('X (meters)')
plt.ylabel('Y (meters)')
plt.legend()
plt.grid(True)
plt.show()
```
